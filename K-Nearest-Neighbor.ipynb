```python
#Import scikit-learn dataset library
from sklearn import datasets

#Load dataset
wine = datasets.load_wine()
```

```python
from sklearn.model_selection import train_test_split

# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(wine.data, wine.target, test_size=0.3) # 70% training and 30% test
print(wine.data)
```

```python
import numpy as np
from statistics import mode

def KNearestNeighbors(k, X_train, y_train, X_predict):
    predictions = [] # stores prediction
    for x in X_predict:
        nearest_neighbors = []
        euclidean = []
        for x_train in X_train:
            euclidean.append(np.sqrt(np.sum(np.power(x - x_train, 2)))) # calculates Euclidean distance for each test point
        sorted_index = np.argsort(euclidean)[0:k] # sorts the distances and stores the indices of the k nearest elements
        for p in sorted_index:
            nearest_neighbors.append(y_train[p]) # appends the classifications of the k nearest elements
        print(nearest_neighbors)
        predictions.append(mode(nearest_neighbors)) # appends the majority classification
    return np.array(predictions)
```

```python
# 1-NN
y_pred = KNearestNeighbors(1, X_train, y_train, X_test)
y_pred_train = KNearestNeighbors(1, X_train, y_train, X_train)

from sklearn import metrics
print("Train Accuracy:",metrics.accuracy_score(y_train, y_pred_train))
print("Test Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
# 7-NN
y_pred = KNearestNeighbors(7, X_train, y_train, X_test)
y_pred_train = KNearestNeighbors(7, X_train, y_train, X_train)

from sklearn import metrics
print("Train Accuracy:",metrics.accuracy_score(y_train, y_pred_train))
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
# 100-NN
y_pred = KNearestNeighbors(100, X_train, y_train, X_test)
y_pred_train = KNearestNeighbors(100, X_train, y_train, X_train)

from sklearn import metrics
print("Train Accuracy:",metrics.accuracy_score(y_train, y_pred_train))
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
# 7-NN with Normalization

from sklearn import preprocessing
scaler = preprocessing.StandardScaler().fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)

y_pred = KNearestNeighbors(7, X_train, y_train, X_test)
y_pred_train = KNearestNeighbors(7, X_train, y_train, X_train)

from sklearn import metrics
print("Train Accuracy:",metrics.accuracy_score(y_train, y_pred_train))
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
